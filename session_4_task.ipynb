###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]
```

````python
# Take input for first list (strings)
L1 = input("Enter elements of List 1 separated by spaces: ").split()

# Take input for second list (strings)
L2 = input("Enter elements of List 2 separated by spaces: ").split()

result = []

# Find the length of the smaller list
min_len = min(len(L1), len(L2))

# Add elements index-wise
for i in range(min_len):
    result.append(L1[i])
    result.append(L2[i])

# Add leftover elements from the longer list
if len(L1) > len(L2):
    for i in range(min_len, len(L1)):
        result.append(L1[i])
else:
    for i in range(min_len, len(L2)):
        result.append(L2[i])

print("New List:", result)
````

### `Problem 2:` Add new item to list after a specified item
Write a program to add item 7000 after 6000 in the following Python List

`Output:`

###`Problem 4:` Running Sum on list
Write a program to print a list after performing running sum on it.

i.e:

`Input:`
```
list1 = [1,2,3,4,5,6]
```
`Output:`
```
[1,3,6,10,15,21]
```

````python
# Write code here

# Take list input from the user
L = list(map(int, input("Enter list elements separated by spaces: ").split()))

running_sum = 0

# Perform running sum
for i in range(len(L)):
    running_sum = running_sum + L[i]
    L[i] = running_sum

print("List after running sum:", L)
````

###`Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
# Write code here

# Take list input from the user
# Take list input from the user
L = list(map(int, input("Enter list elements separated by spaces: ").split()))

result = []

# Calculate sum of current element and all greater elements
for i in range(len(L)):
    total = L[i]

    for j in range(len(L)):
        if L[j] > L[i]:
            total += L[j]

    result.append(total)

print("Result List:", result)
````

###`Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
# Write code here

# Take input for the first list
L1 = list(map(int, input("Enter elements of List 1 separated by spaces: ").split()))

# Take input for the second list
L2 = list(map(int, input("Enter elements of List 2 separated by spaces: ").split()))

common = []

# Find common unique elements
for item in L1:
    if item in L2 and item not in common:
        common.append(item)

# Sort the common elements
common.sort()

print("Common unique items in increasing order:", common)
````

###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
# Write code here

# Function to calculate the product of numeric digits in a string
def product_value(s):
    product = 1
    has_digit = False

    for ch in s:
        if ch.isdigit():
            product *= int(ch)
            has_digit = True

    if has_digit:
        return product
    else:
        return 1

# Take input from the user
L = input("Enter alphanumeric strings separated by spaces: ").split()

# Sort the list based on the product value
L.sort(key=product_value)

print("Sorted List:", L)
````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# Code here
# Take input from the user
text = input("Enter a string: ")

result = ""

for ch in text:
    if ch.isupper() and result:
        result += " "
    result += ch

print("Output:", result)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
# Write code here

# Take input for the first list
L1 = list(map(int, input("Enter elements of List 1 separated by spaces: ").split()))

# Take input for the second list
L2 = list(map(int, input("Enter elements of List 2 separated by spaces: ").split()))

union = []

# Add unique elements from the first list
for item in L1:
    if item not in union:
        union.append(item)

# Add unique elements from the second list
for item in L2:
    if item not in union:
        union.append(item)

print("Union of the two lists:", union)
````

### `Problem 12:` Write a program that can find the max number of each row of a matrix

**Example:**

Input:

```bash
[[1,2,3],[4,5,6],[7,8,9]]
```

Output:
```bash
[3,6,9]
```

````python
# Write code here

# Take number of rows and columns
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

matrix = []

# Input the matrix
print("Enter the matrix elements:")
for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

# Find the maximum element of each row
print("Maximum element in each row:")

for i in range(rows):
    max_num = matrix[i][0]

    for j in range(1, cols):
        if matrix[i][j] > max_num:
            max_num = matrix[i][j]

    print(max_num)
````

````python
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

matrix = []

for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

print("Maximum element in each row:")

for row in matrix:
    print(max(row))
````

### `Problem 14:` Write a list comprehension that can transpose a given matrix

matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

[1, 4, 7]<br>
[2, 5, 8]<br>
[3, 6, 9]<br>

````python
# Write code here
# Given matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Transpose using list comprehension
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

# Print the transposed matrix
for row in transpose:
    print(row)
````

````python
# Input matrix dimensions
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

print("Enter the matrix:")
for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

# Transpose using list comprehension
transpose = [[row[i] for row in matrix] for i in range(cols)]

print("Transposed Matrix:")
for row in transpose:
    print(row)
````

### `Problem 15:` Write a list comprehension that can flatten a nested list

Input<br>
matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

Output:<br>
[1, 2, 3, 4, 5, 6, 7, 8, 9]

````python
# Write code here
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

matrix = []

print("Enter the matrix:")
for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

# Flatten using list comprehension
flattened = [element for row in matrix for element in row]

print("Flattened List:", flattened)
````

````python
# Given matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Flatten the matrix using list comprehension
flattened = [element for row in matrix for element in row]

print(flattened)
````
